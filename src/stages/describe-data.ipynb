### Describe

```python
df = pd.read_json(f"{directory}/colect-data.json")
df.head()
```

```python
num_registros = len(df)
print(f"Número total de registros: {num_registros}")
```

```python
categories = ['TIPO_INGRESSO', 'CAMPUS', 'CURSO', 'NATURALIDADE', 'UF_NATURALIDADE',
               'PAIS_NATURALIDADE', 'CIDADE_ENDERECO', 'UF_ENDERECO', 'SEXO', 'COTA',
               'TIPO_COTA', 'ESCOLARIDADE_PAI', 'ESCOLARIDADE_MAE', 'DEFICIENCIA']

for column in categories:
    print(f"\nAnálise de {column}:")
    print(df[column].value_counts(normalize=True))
```

```python
ingresso = ['2020', '2021', '2022', '2023']

numericas = ['ANO_INGRESSO', 'ANO_NASC']

for ano in ingresso:
    print(f"ANO_INGRESSO para o ano {ano}")
    print(f"Moda do campo ANO_NASC para o ano {ano}: {df[df['ANO_INGRESSO'] == int(ano)]['ANO_NASC'].mode()[0]}")
```

```python
print(f"\nNúmero de ID_DISCENTE únicos: {df['ID_DISCENTE'].nunique()}")
print(f"Número de COD_CURSO únicos: {df['COD_CURSO'].nunique()}")

top_bairros = df['BAIRRO_ENDERECO'].value_counts().head(10)
print("\nTop 10 Bairros mais frequentes:")
print(top_bairros)
```

```python
print("\nDados ausentes por coluna:")
print(df.isnull().sum())
```

```python
transicoes = df.groupby(['CIDADE_ENDERECO', 'CAMPUS']).size().unstack(fill_value=0)

stochastic_matrix = transicoes.div(transicoes.sum(axis=1), axis=0)

print("Matriz Estocástica de Transição:")
print(stochastic_matrix)
```
